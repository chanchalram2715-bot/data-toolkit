2. How does broadcasting work in NumPy?
 - Broadcasting lets NumPy do math on arrays of different shapes without making them the same size manually.

Think of it like stretching the smaller array to match the bigger one automatically.

 - How it Works (Simple Rules)

Look at the shapes of the arrays from right to left.

Two dimensions are compatible if:

They are equal, or

One of them is 1 (then it is stretched).

If not, NumPy will give an error.

 - Key Idea

Broadcasting = automatic stretching of arrays to do math together.

Saves time, avoids loops, and makes your code faster.

4. Explain the use of the groupby() method in Pandas
 - he groupby() method in Pandas is used to split data into groups based on some criteria.

After grouping, you can perform operations on each group separately, like sum, mean, count, etc.

Think of it as “divide and conquer” for your data.

- How it Works

groupby() works in three steps (commonly called the split-apply-combine method):

Split – Divide the data into groups based on a column.

Apply – Apply a function (sum, mean, count, etc.) to each group.

Combine – Combine the results into a new DataFrame or Series.

17. Explain the difference between apply() and map() in Pandas
 - map()

Works only on a Series (single column).

Applies a function element-wise to each value.

Can also map using a dictionary or another Series.

Returns a Series of the same length.

Best for simple element-wise operations.

 - apply()

Works on a Series or DataFrame.

Applies a function element-wise (Series) or row/column-wise (DataFrame).

Returns a Series, DataFrame, or scalar depending on the function.

More flexible for complex operations across multiple columns or rows.

Best for aggregations, custom functions, or operations on rows/columns.

21. Why is NumPy’s array slicing faster than Python’s list slicing
 - Contiguous Memory Storage

NumPy arrays are stored in contiguous blocks of memory, unlike Python lists (which store references to objects).

This allows faster access and iteration over elements.

 - Homogeneous Data Type

All elements in a NumPy array are of the same type, enabling optimized low-level computations.

Python lists can hold mixed types, which adds overhead.

 - Vectorized Operations

Slicing in NumPy supports view-based operations, not copies (by default).

This reduces memory usage and speeds up operations.

 - Optimized C Implementation

NumPy is implemented in C, so slicing and operations run at compiled code speed, not Python’s interpreted speed.

 - Reduced Overhead

Python lists require type checking and object handling for each element, which slows down slicing.

NumPy avoids this by operating directly on raw memory.

```python
#11. Perform matrix multiplication using NumPy

import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Matrix B (3x2)
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

result = np.dot(A, B)
print("Matrix multiplication result using np.dot:\n", result)

# Using @ operator
result2 = A @ B
print("Matrix multiplication result using @:\n", result2)
```
